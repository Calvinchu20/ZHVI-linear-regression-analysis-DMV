```python
import pandas as pd 
import datetime 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HomeZHVI.csv')
```

```python
df
```

```python
df_MD = df[df['StateName'] =='MD']
df_MD
```

```python
"Virgina dataset has DC in it "
df_VA = df[df['StateName'] =='VA']
df_VA
```

```python
df_DMV = pd.concat([df_MD,df_VA])
df_DMV
```

```python
"remove unnecessary columns such as RegionType"
df_DMV = df_DMV.drop(columns=['RegionType','RegionID'])
df_DMV.fillna(0)
```

```python
"most recent month is 8-31-2024"
"only filter the data to have this date 8-31 from 2019-2024(most recent)"


df_final = df_DMV[['SizeRank','RegionName','StateName','2019-08-31','2020-08-31','2021-08-31','2022-08-31','2023-08-31','2024-08-31']]
df_final
df_DMV = df_final.round(2)
df_DMV
```

```python
x = df_DMV['RegionName']
y= df_DMV["2024-08-31"]
plt.bar(x,y)
plt.xlabel("DMV Cities")
plt.ylabel("ZHVI housing value")
plt.title("The ZHVI housing value of all major cities in DMV(8-31-2024)")
plt.xticks(rotation = 90)
plt.show()
```

```python
"convert the date format into only years"
import datetime 
for col in df_DMV.columns:
    if col == 'SizeRank' or col =='RegionName' or col == 'StateName':
        continue
   
    date = datetime.datetime.strptime(col,'%Y-%m-%d')
    df_DMV = df_DMV.rename(columns={col:date.year})

```

```python
"line graph progression of ZHVI home value from 2019-2024 of most recent record(8-31-2024)"
import seaborn as sns
x = df_linear['Year']
y = df_linear['HousingValue']
plt.scatter(x,y)
plt.title("Progression of ZHVI home value in the DMV from 2019 - 8-31-2024")
plt.xlabel("Years")
plt.ylabel("ZHVI home value")
```

Data ML model: regression of years and housing value ZHVI to established a relationship and predict future ZHVI in the next 5 years

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X = df_linear["Year"]
y = df_linear['HousingValue']
```

```python
x_train,x_test,y_train,y_test  = train_test_split(x,y, test_size=0.2)
```

```python
x_train = x_train.values.reshape(-1,1)
y_train = y_train.values.reshape(-1,1)
model = linear_model.LinearRegression()
model.fit(x_train,y_train)
```

```python
x_test = x_test.values.reshape(-1,1)
y_predict = model.predict(x_test)
```

```python
mean_squared_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```

```python
plt.figure(figsize=(10, 6))
x = df_linear[['Year']]
y = df_linear['HousingValue']
plt.scatter(x_train,y_train)
plt.title("Progression of ZHVI home value in the DMV from 2019 - 8-31-2024")
plt.xlabel("Years")
plt.ylabel("ZHVI home value")
plt.legend()
plt.grid(True)
y_pred = model.predict(x_train)
plt.plot(x_train,y_pred,color ="red")
plt.show()
```
